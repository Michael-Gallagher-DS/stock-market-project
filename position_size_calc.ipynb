```python
# Import tkinter and customtkinter libraries
from tkinter import *
import customtkinter
```

```python
#Initiate the root
app = customtkinter.CTk()
app.geometry("310x340")
app.title("Position Sizing Calculator")
```

```python
#Set default portfolio value and risk %
portfolio_text = customtkinter.StringVar()
portfolio_text.set("100000")
risk_text = customtkinter.StringVar()
risk_text.set("0.01")
```

```python
#Create portfolio size and risk entry fields
portfolio_size = customtkinter.CTkEntry(master=app, textvariable=portfolio_text,placeholder_text="Portfolio Size", width=70, height=15)
portfolio_size.grid(row=0, column=0, stick=W, padx=5, pady=5)

risk_percent = customtkinter.CTkEntry(master=app, textvariable=risk_text,placeholder_text="Risk %", width=70, height=15)
risk_percent.grid(row=1, column=0, stick=W, padx=5, pady=5)
```

```python
# Create button to calculate allowed risk
# Initialize risk_amt_label as None
risk_amt_label = None

def risk_amount_calculation():
    global risk_amt_label
    try:
        # Get values from entry fields
        portfolio_size_val = float(portfolio_size.get())
        risk_val = float(risk_percent.get())

        # Perform multiplication
        result = portfolio_size_val * risk_val

        # If risk_amt_label is not None, destroy it
        if risk_amt_label:
            risk_amt_label.destroy()

        # Create and place a new label
        risk_amt_label = customtkinter.CTkLabel(app, text=f"Risk Amount: ${result}", fg_color="transparent", width=20, height=10)
        risk_amt_label.grid(row=1, column=1, stick=W, padx=5, pady=5)
    except ValueError:
        # Handle the case when the input is not a valid number
        if risk_amt_label:
            risk_amt_label.destroy()
        risk_amt_label = customtkinter.CTkLabel(app, text="Invalid input.", fg_color="transparent", width=20, height=10)
        risk_amt_label.grid(row=1, column=1, stick=W, padx=5, pady=5)
```

```python
#Create entry fields for stock price, target and stop loss
entry_price = customtkinter.CTkEntry(master=app, placeholder_text="Entry", width=70, height=15)
entry_price.grid(row=4, column=0, stick=W, padx=5, pady=5)

stop_price = customtkinter.CTkEntry(master=app, placeholder_text="Stop", width=70, height=10)
stop_price.grid(row=5, column=0, stick=W, padx=5, pady=5)

target_price = customtkinter.CTkEntry(master=app, placeholder_text="Target", width=70, height=10)
target_price.grid(row=6, column=0, stick=W, padx=5, pady=5)
```

Create a function to calculate the profit per share.

```python
#Per share calculations
def per_share_calculation():
    # Get values from entry fields
    entry_price_val = float(entry_price.get())
    target_price_val = float(target_price.get())

    # Perform calculation for short vs long
    if entry_price_val < target_price_val:
        profit_share_result = target_price_val - entry_price_val
    else:
        profit_share_result = abs(entry_price_val - target_price_val)




    # Get values from entry fields
    stop_price_val = float(stop_price.get())

    # Perform calculation for short vs long
    if entry_price_val > stop_price_val:
        risk_share_result = entry_price_val - stop_price_val
    else:
        risk_share_result = abs(entry_price_val - stop_price_val)


    



    # Perform calculation for profit/share
    if entry_price_val < target_price_val:
        profit_share_result = target_price_val - entry_price_val
    else:
        profit_share_result = abs(entry_price_val - target_price_val)
    
    # Perform calculation for risk/share
    if entry_price_val > stop_price_val:
        risk_share_result = entry_price_val - stop_price_val
    else:
        risk_share_result = abs(entry_price_val - stop_price_val)

    rr_ratio_result = profit_share_result/risk_share_result

    # Create and place the RR label
    rr_ratio_label = customtkinter.CTkLabel(app, text=f"R/R: {round(rr_ratio_result,1)}", fg_color="transparent")
    rr_ratio_label.grid(row=10, column=0, stick=W, padx=5, pady=5)

```

Create function for number of shares to purchase.

```python
def num_shares_calculation():
    # Get values from entry fields
    entry_price_val = float(entry_price.get())
    stop_price_val = float(stop_price.get())

    # Perform calculation for short vs long.
    if entry_price_val > stop_price_val:
        risk_share_result = entry_price_val - stop_price_val
    else:
        risk_share_result = abs(entry_price_val - stop_price_val)

    # Get values from entry fields to find the total shares you're able to trade.
    portfolio_size_val = float(portfolio_size.get())
    risk_val = float(risk_percent.get())

    # Perform multiplication
    total_avail_risk_result = portfolio_size_val * risk_val
    num_shares_result = total_avail_risk_result/risk_share_result
    
    if (num_shares_result * entry_price_val) > portfolio_size_val:
        num_shares_result = portfolio_size_val / entry_price_val
    else:
        num_shares_result = total_avail_risk_result/risk_share_result

    # Create total cost of trade, profit and risk
    total_cost_of_trade = num_shares_result * entry_price_val




    
    # Calculate total profit of trade
    target_price_val = float(target_price.get())

    # Perform calculation for short vs long
    if entry_price_val < target_price_val:
        profit_share_result = target_price_val - entry_price_val
    else:
        profit_share_result = abs(entry_price_val - target_price_val)

    total_profit_of_trade = num_shares_result * profit_share_result
    




    # Perform calculation for short vs long
    if entry_price_val > stop_price_val:
        risk_share_result = entry_price_val - stop_price_val
    else:
        risk_share_result = abs(entry_price_val - stop_price_val)

    total_risk_of_trade = num_shares_result * risk_share_result


    num_shares_rounded = round(num_shares_result, 0)
    num_shares_formatted = f"{num_shares_rounded:,.0f}"
    num_shares_result_label = customtkinter.CTkLabel(app, text=f"Shares: {num_shares_formatted}", fg_color="transparent")

    num_shares_result_label.grid(row=8, column=0, stick=W, padx=5, pady=5)



    # Fixing total_cost_of_trade_label
    total_cost_of_trade_rounded = round(total_cost_of_trade, 2)
    total_cost_of_trade_formatted = f"{total_cost_of_trade_rounded:,.2f}"
    total_cost_of_trade_label = customtkinter.CTkLabel(app, text=f"Total Cost: ${total_cost_of_trade_formatted}", fg_color="transparent")

    total_cost_of_trade_label.grid(row=8, column=1, stick=W, padx=5, pady=5)



    # Fixing total_profit_of_trade_label
    total_profit_of_trade_rounded = round(total_profit_of_trade, 2)
    total_profit_of_trade_formatted = f"{total_profit_of_trade_rounded:,.2f}"
    total_profit_of_trade_label = customtkinter.CTkLabel(app, text=f"Profit: ${total_profit_of_trade_formatted}", fg_color="transparent")

    total_profit_of_trade_label.grid(row=9, column=1, stick=W, padx=5, pady=5)



    # Fixing total_risk_of_trade_label
    total_risk_of_trade_rounded = round(total_risk_of_trade, 2)
    total_risk_of_trade_formatted = f"{total_risk_of_trade_rounded:,.2f}"
    total_risk_of_trade_label = customtkinter.CTkLabel(app, text=f"Risk: ${total_risk_of_trade_formatted}", fg_color="transparent")

    total_risk_of_trade_label.grid(row=9, column=0, stick=W, padx=5, pady=5)
```
